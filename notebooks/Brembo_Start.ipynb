```python
def plot(cols):
  idx = 0
  for index, row in dfs.iterrows():
    if idx == 10:
      break


    index = [ i for i in range(31) ]

    dft = {}

    for col in cols:
      min_ = row[col].min()
      max_ = row[col].max()
      row[col] = (row[col] - min_)/(max_ - min_)
      dft[col] = row[col]

    df_plot = pd.DataFrame(dft, index=row["time"])
    print(df_plot.corr())
    df_plot.plot.line()

    idx += 1
```

```python
plot(["pressure_bar", "temp_c", "mu", "speed_kph"])
```

```python
def is_valid(composition):
  material_classes = [ 'A', 'B', 'C', 'D', 'E', 'F' ]
  bounds = {
      'A': [0, 12],
      'B': [1, 30],
      'C': [0, 18],
      'D': [0.4, 1],
      'E': [45, 92],
      'F': [3, 27]
  }

  amounts = {}
  for material, value in composition.items():
    material_class = material.split('_')[0]
    if material_class in amounts:
      amounts[material_class] += value
    else:
      amounts[material_class] = value

  if sum(amounts.values()) != 100:
    return False
  for material_class in material_classes:
    if amounts[material_class] < bounds[material_class][0] or amounts[material_class] > bounds[material_class][1]:
      return False

  return True
```

```python
is_valid({
    'A_1': 1.0,
    'B_2': 0,
    'C_5': 3,
})
```
